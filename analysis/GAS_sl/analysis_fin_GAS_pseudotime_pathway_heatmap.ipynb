# Merge 8 pathway analysis results

```python
work_dir_filename='/Users/wanglab/Documents/ShengyuLi/Velocity/data/Gastrulation/velocity_result/result_detailcsv/2000_genes_v2_smooth_bin/2000genes_with_bin_and_smooth/pseudotime/radis/rsquare_point_eight_not_seperate_no_HP_cls/reorder_david/david_count0_ease0/david_cluster'

def merge_david_result(work_dir_filename,para,method):
    merged_david_list=[]
    for i in range(1,9):
        david_result=pd.read_table((work_dir_filename+str(i)+'.txt'))
        david_result=david_result[['Term',para]]
        david_result.columns=['Term',(para+'_cluster'+str(i))]
        merged_david_list.append(david_result)

        if method=='outer':
            david_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Term'],
                                                    how='outer'), merged_david_list).fillna(1)
        elif method=='inner':
            david_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Term'],
                                                    how='inner'), merged_david_list)
    return(david_merged)
```

```python
para='PValue'
method='outer'
david_merged_outer=merge_david_result(work_dir_filename,para,method)
david_merged_outer.to_csv('/Users/wanglab/Documents/ShengyuLi/Velocity/data/Gastrulation/velocity_result/result_detailcsv/2000_genes_v2_smooth_bin/2000genes_with_bin_and_smooth/pseudotime/radis/rsquare_point_eight_not_seperate_no_HP_cls/reorder_david/david_count0_ease0/david_merged_P_outer.csv',index=False)

# para='PValue'
# method='inner'
# david_merged_inner = merge_david_result(work_dir_filename,para,method)
# david_merged_inner.to_csv('/Users/wanglab/Documents/ShengyuLi/Velocity/data/Gastrulation/velocity_result/result_detailcsv/2000_genes_v2_smooth_bin/2000genes_with_bin_and_smooth/pseudotime/radis/rsquare_point_eight_not_seperate_no_HP_cls/reorder_david/david_count0_ease0/david_merged_P_inner.csv',index=False)
# david_merged_inner

david_merged_outer
```

# Filter pathways of p=1 in all 8 clusters

```python
def filter_para_all_p1(df):
    # filter pathways of p=1 in all clusters
    sum_column=df[df.columns[1:]].sum(axis=1)
    sum_column=pd.DataFrame(sum_column,columns=['sum_col'])
    sum_column_eight_idx=sum_column[sum_column.sum_col==8].index
    sum_column_eight_idx
    filtered_df=df[~df.index.isin(sum_column_eight_idx)]

    return(filtered_df)

david_merged_outer_filtered=filter_para_all_p1(david_merged_outer)
david_merged_outer_filtered.to_csv('/Users/wanglab/Documents/ShengyuLi/Velocity/data/Gastrulation/velocity_result/result_detailcsv/2000_genes_v2_smooth_bin/2000genes_with_bin_and_smooth/pseudotime/radis/rsquare_point_eight_not_seperate_no_HP_cls/reorder_david/david_count0_ease0/david_merged_P_outer_filtered.csv',index=False)
```
