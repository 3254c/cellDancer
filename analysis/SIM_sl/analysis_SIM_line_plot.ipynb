```python
def brief_line_plot(brief,interval=1, save_path=None,name=None):
    plt.figure()
    brief=brief.copy()
    brief_pivot=brief.pivot(index='gene_name', columns='epoch', values='cost')

    brief_pivot

    brief_pivot_max=brief_pivot.max()[::interval]
    brief_pivot_min=brief_pivot.min()[::interval]
    plt.fill_between(brief_pivot_max.index,brief_pivot_min,brief_pivot_max,color='#B1B1B1')
    
    brief_pivot_quarter1=brief_pivot.quantile(0.1)[::interval]
    brief_pivot_quarter2=brief_pivot.quantile(0.9)[::interval]
    plt.fill_between(brief_pivot_max.index,brief_pivot_quarter1,brief_pivot_quarter2,color='#7E7E7E')
    
    # brief_pivot_quarter3=brief_pivot.quantile(0.2)[::interval]
    # brief_pivot_quarter4=brief_pivot.quantile(0.8)[::interval]
    # plt.fill_between(brief_pivot_max.index,brief_pivot_quarter3,brief_pivot_quarter4,color='yellow')
    
    
    brief_pivot_quarter5=brief_pivot.quantile(0.4)[::interval]
    brief_pivot_quarter6=brief_pivot.quantile(0.6)[::interval]
    plt.fill_between(brief_pivot_max.index,brief_pivot_quarter5,brief_pivot_quarter6,color='black')
    
    plt.ylim(0,0.2)
    plt.savefig(save_path+name)
    
    
```

```python
save_path='/Users/wanglab/Documents/ShengyuLi/Velocity/data/simulation/data/alalysis/line_cost/'
brief_line_plot(brief_normal,interval=1,save_path=save_path,name='cost_line_normal.pdf')
brief_line_plot(brief_multi,interval=1,save_path=save_path,name='cost_line_multi.pdf')
brief_line_plot(brief_wing,interval=1,save_path=save_path,name='cost_line_wing.pdf')
brief_line_plot(brief_back,interval=1,save_path=save_path,name='cost_line_back.pdf')
```
